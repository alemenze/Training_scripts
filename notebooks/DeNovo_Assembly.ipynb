## Raw data download

```python
%%bash
fastq-dump --split-files --gzip --outdir ./raw/ SAMN04216069
```

## FASTQC

```python
%%bash
fastqc raw/SAMN04216069_1.fastq.gz
fastqc raw/SAMN04216069_2.fastq.gz
```

## Trimmomatic

```python
%%bash
TrimmomaticPE raw/SAMN04216069_1.fastq.gz raw/SAMN04216069_1.fastq.gz \
            raw/SAMN04216069_1_paired.fastq.gz raw/SAMN04216069_1_unpaired.fastq.gz \
            raw/SAMN04216069_2_paired.fastq.gz raw/SAMN04216069_1_unpaired.fastq.gz \
            LEADING:20 TRAILING:20 SLIDINGWINDOW:5:20 \
            MINLEN:140 

fastqc raw/SAMN04216069_1_paired.fastq.gz
fastqc raw/SAMN04216069_2_paired.fastq.gz                    
multiqc .
```

## SPAdes

```python
%%bash
./SPAdes-3.14.0-Linux/bin/spades.py -1 raw/SAMN04216069_1_paired.fastq.gz -2 raw/SAMN04216069_2_paired.fastq.gz \
          -o spades-default-assembly/ --only-assembler

./SPAdes-3.14.0-Linux/bin/spades.py -1 raw/SAMN04216069_1_paired.fastq.gz -2 raw/SAMN04216069_2_paired.fastq.gz \
          -o spades-errorcorrected-assembly/

./SPAdes-3.14.0-Linux/bin/spades.py -1 raw/SAMN04216069_1_paired.fastq.gz -2 raw/SAMN04216069_2_paired.fastq.gz \
          -o spades-careful-assembly/ --careful
```

## Quast

```python
%%bash
quast.py -o quast_output -b \
    -l "SPAdes-default, SPAdes-error-corrected, SPAdes-careful" \
    spades-default-assembly/contigs.fasta \
    spades-errorcorrected-assembly/contigs.fasta \
    spades-careful-assembly/contigs.fasta \
```

## Prokka

```python
%%bash
docker run --rm=True -v $PWD:/data -u $(id -u):$(id -g) staphb/prokka:latest prokka \
            --outdir prokka_output --centre MAGIC --compliant spades-careful-assembly/contigs.fasta
```
